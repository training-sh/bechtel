```python
from openai import OpenAI
import os

DATABRICKS_TOKEN = "" # os.environ.get("DATABRICKS_TOKEN")

client = OpenAI(
  api_key=DATABRICKS_TOKEN, # your personal access token
  base_url='https://xxxxxxx.cloud.databricks.com/serving-endpoints', # your Databricks workspace instance
)

chat_completion = client.chat.completions.create(
  messages=[
    {
      "role": "system",
      "content": "You are an AI assistant",
    },
    {
      "role": "user",
      "content": "What is a mixture of experts model?",
    }
  ],
  model="databricks-meta-llama-3-1-8b-instruct",
  max_tokens=256
)

print(chat_completion.choices[0].message.content)
```

```python
import tiktoken

"""
The LLM cannot understand text directly.
A tokenizer is simply a translator.

It converts text into numbers.
"""

enc = tiktoken.get_encoding("cl100k_base")

text = "Who invented Python?"

tokens = enc.encode(text)

print(text)
print(tokens)

for t in tokens:
    print(t, enc.decode([t]))
```
