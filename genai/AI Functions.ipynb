```python
%sql
SELECT
ticket_id,

ai_query(
  'databricks-gpt-oss-20b',
  CONCAT(
    'Classify complaint category from: ',
    complaint_text
  )
) AS complaint_category

FROM complaints.bronze_tickets;
```

```python
%sql
SELECT
ticket_id,

ai_summarize(complaint_text) AS complaint_summary

FROM complaints.bronze_tickets;
```

```python
%sql
-- severity

SELECT
ticket_id,

ai_query(
'databricks-gpt-oss-20b',
CONCAT(
'Classify severity as
Low, Medium, High, Critical.

Complaint:
',
complaint_text,
'
Notes:
',
technician_notes
)
) AS severity

FROM complaints.bronze_tickets;
```

```python
%sql
-- Sentiment
SELECT
ticket_id,

ai_analyze_sentiment(
complaint_text
) AS sentiment

FROM complaints.bronze_tickets;
```

```python
%sql
CREATE OR REPLACE TABLE complaints.silver_tickets AS

SELECT

ticket_id,
product_name,
complaint_text,

ai_summarize(complaint_text)
    AS complaint_summary,

ai_analyze_sentiment(complaint_text)
    AS sentiment,

ai_query(
'databricks-gpt-oss-20b',
CONCAT(
'Classify complaint category: ',
complaint_text
))
AS complaint_category,

CURRENT_TIMESTAMP()
AS processed_timestamp

FROM complaints.bronze_tickets;
```
